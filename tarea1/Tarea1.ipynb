## 1.- Construcción


### Método de la Bisección.

La función $f$ a utilizar es:

\begin{equation*}
 f(x_i, x_{i+1}, y_i, y_{i+1}) =  y_i  -y_{i+1}  + \dfrac{x_{i+1}-x_i}{2}\left(y_{i+1}\cos{(x_{i+1}+y_{i+1})}+y_i\cos{(x_i+y_i)}\right)\\
\end{equation*}

Dicho función nace de despejar la ecuación, de modo que quede igualada a cero.

El intervalo a utilizar es:

\begin{equation*}
\left[\frac{y_i + \frac{x_{i+1} - x_{i}}{2} y_i\cos{(x_i+y_i)}}{1+\frac{x_{i+1} - x_{i}}{2}}, \frac{y_i+\frac{x_{i+1} - x_{i}}{2} y_i\cos{(x_i+y_i)}}{1-\frac{x_{i+1} - x_{i}}{2}}\right]
\end{equation*}

Este intervalo nace de acotar la función $f$ de la siguiente forma:

Para la iteración $i$, los valores $x_i$, $y_i$, $x_{i+1}$ son conocidos y ademas $y_{i+1}$ es la variable desconocida. En base a esto, las expresiones que no dependen de esta variable las podemos considerar constantes para este analisis.

Sea:
\begin{equation*}
A = y_i \\
B = \frac{x_{i+1} - x_{i}}{2} \\
C = x_{i+1} \\
D = y_i \cos{(x_i + y_i)} \\
x = y_{i+1}
\end{equation*}

De modo que:
\begin{equation*}
f(x) =  A - x + B\left(x\cos{(C + x)} + D\right)\\
\end{equation*}

Ademas sabemos que:
\begin{equation*}
-1 \leq \cos(x) \leq 1 \\
-1 \leq x\cos(C + x) \leq 1 \\
-x \leq x\cos(C + x) \leq x \\
-x + D \leq x\cos(C + x) + D \leq x + D \\
B(-x + D) \leq B(x\cos(C + x) + D) \leq B(x + D) \\
A - x + B(-x + D) \leq A - x + B(x\cos(C + x) + D) \leq A - x + B(x + D) \\
A - x + B(-x + D) \leq f(x) \leq A - x + B(x + D) \\
\end{equation*}

Por lo que la funcion $f(x)$ esta acotada por $A + DB - x(1 + B)$ y $A + DB + x(B-1)$. Ademas, queremos poder acotar por el teorema del sandwich al valor 0, igualamos estas ecuaciones a cero:
\begin{align*}
A + DB - x(1 + B) &= 0 \\
x(1 + B) &= A + DB \\
x &= \frac{A + DB}{1 + B} \\
\end{align*}
y
\begin{align*}
A + DB + x(B - 1) &= 0 \\
x(B - 1) &= -(A + DB) \\
x &= \frac{(A + DB)}{1 - B} \\
\end{align*}

Por lo que el intervalo esta dado por $\left[\frac{A + DB}{1 + B}, \frac{A + DB}{1 - B}\right]$


### Iteración de Punto Fijo

\begin{equation*}
 g(x_i, x_{i+1}, y_i, y_{i+1}) =  y_i + \dfrac{x_{i+1}-x_i}{2}\left(y_{i+1}\cos{(x_{i+1}+y_{i+1})}+y_i\cos{(x_i+y_i)}\right)\\
\end{equation*}

Esta funcion nace de
\begin{equation*}
f(x_i, x_{i+1}, y_i, y_{i+1}) =  0 \\
y_i  -y_{i+1}  + \dfrac{x_{i+1}-x_i}{2}\left(y_{i+1}\cos{(x_{i+1}+y_{i+1})}+y_i\cos{(x_i+y_i)}\right) + y_{i+1} =  y_{i+1} \\
y_i + \dfrac{x_{i+1}-x_i}{2}\left(y_{i+1}\cos{(x_{i+1}+y_{i+1})}+y_i\cos{(x_i+y_i)}\right) =  y_{i+1} \\
g(x_i, x_{i+1}, y_i, y_{i+1})=  y_{i+1} \\
\end{equation*}


El _initial guess_ del punto fijo fue seleccionado arbitrariamente como 1. Un indicador para decidir este _initial guess_ fue que el $y_0$ es igual a 1. <!--Y porque fue el primero que funciono...-->

### Método de Newton. 

\begin{equation*}
f(x_i, x_{i+1}, y_i, y_{i+1}) =  y_i  -y_{i+1}  + \dfrac{x_{i+1}-x_i}{2}\left(y_{i+1}\cos{(x_{i+1}+y_{i+1})}+y_i\cos{(x_i+y_i)}\right)\\
\frac{df}{d y_{i+1}}(x_i, x_{i+1}, y_i, y_{i+1}) = \frac{x_{i+1} - x_i}{2} (y_{i+1} \sin{(x_{i+1} + y_{i+1})} + \cos{(x_{i+1} + y_{i+1})}) - 1
\end{equation*}

initial guess = $\frac{y_i+\frac{x_{i+1} - x_{i}}{2} y_i\cos{(x_i+y_i)}}{1-\frac{x_{i+1} - x_{i}}{2}}$

Como sabemos de la explicacion de la biseccion, la funcion en cuestion esta acotada por 2 rectas. Aprovecharemos esto para usar uno de esos valores como initial guess.


## 2.- Implementación
Aquí usted debe implementar cada uno de los algoritmos propuestos en la pregunta anterior de la forma que los explicó. 
Usted tendrá que considerar un presupuesto $N\_budget$ de *operaciones elementales*, es decir, cada vez que en su implementación realice una operación elemental la debe descontar de su presupuesto y el algoritmo debe retornar el vector $\mathbf{y}$ de soluciones encontradas hasta donde pudo obtener una solución.
Si el presupuesto es menor a lo que necesita para poder obtener las $M\_$intervals raices, el vector debe retornar NaN en las componentes que no pudo obtener. Condicionales no se consideran operaciones elementales pero si hubieran operaciones elementales dentro de un condicional debén incluirse y descontarse del presupuesto cada vez que se realicen.
El costo de cada operación viene dado por la siguiente tabla: <br> 

|Operaciones elementales|Costo|
|:----------:|:-------------:|
| +,-,*,/ |  1 |
| Funciones Elementales (trig, etc) |    50   |


```python
'''
Funcion f() que se usara en la busqueda de ceros para el metodo de la biseccion y newton.
Output:
y - (tuple) La funcion evaluada con los parametros entregados y el costo computacional de evaluar todos estos valores.
'''
def f_aveng(x_i, x_i_1, y_i, y_i_1):
    op_cost = 50*2 + 9 # 2 cos && 9 ops
    return y_i - y_i_1 + (x_i_1 - x_i)/2 * (y_i_1 * np.cos(x_i_1 + y_i_1) + y_i * np.cos(x_i+y_i)), op_cost

'''
Funcion g() que se usara en el metodo de punto fijo.
Output:
y - (tuple) La funcion evaluada con los parametros entregados y el costo computacional de evaluar todos estos valores.
'''
def g_aveng(x_i, x_i_1, y_i, y_i_1):
    op_cost = 50*2 + 8 # 2 cos && 8 ops
    return y_i + (x_i_1 - x_i)/2 * (y_i_1 * np.cos(x_i_1 + y_i_1) + y_i * np.cos(x_i+y_i)), op_cost

'''
La derivada de la funcion f(). Se usara en el metodo de newton.
Output:
y - (tuple) La funcion evaluada con los parametros entregados y el costo computacional de evaluar todos estos valores.
'''
def d_f_aveng(x_i, x_i_1, y_i, y_i_1):
    op_cost = 50*2 + 10 # sin + cos + 8 ops
    return (x_i_1 - x_i)/2 * (y_i_1 * np.sin(x_i_1 + y_i_1) + np.cos(x_i_1 + y_i_1)) - 1, op_cost
```

```python
'''
Metodo de la biseccion.
Input:
x_i, x_i_1, y_i - (double) Parametros a evaluar en la funcion, los cuales ya son conocidos en la iteracion.
a, b - (double) El rango de los initial guess a evaluar.
tol - (double) La tolerancia del error.
N_budget - (integer) El presupuesto que nos queda para ejecutar este metodo.
Output:
y - (tuple) El valor y_i_1 que vuelve cero la ecuacion y lo que resta de presupuesto.
'''
def bisect(x_i, x_i_1, y_i, a, b, tol, N_budget):
    def f(val):
        return f_aveng(x_i, x_i_1, y_i, val)
    # Se verifica que los intial guess nos den un rango logico.
    f_a, aux = f(a)
    N_budget -= aux
    f_b, aux = f(b)
    N_budget -= aux - 1
    mult = f(a)[0]*f(b)[0]
    if mult > 0 or N_budget < 0:
        return np.nan, -1
    elif mult == 0: # Quizas un initial guess nos da el resultado buscado.
        if f_a == 0:
            return a, N_budget
        else:
            return b, N_budget

    c = (b+a)/2; N_budget -= 2
    while True:
        N_budget -= 3
        if not (np.abs(b-a)/2 > tol):
            break
        f_c, aux = f(c)
        N_budget -= aux
        if f_c == 0:
            break
        f_a, aux = f(a)
        N_budget -= aux - 1
        if f_a*f_c < 0:
            b = c
        else:
            a = c
        N_budget -= 2
        c = (b+a)/2
        if N_budget < 0:
            break
    return c, N_budget

'''
Calcula el a y el b para el metodo de la biseccion.
Input:
x_i, x_i_1, y_i - (double) Parametros a evaluar en la funcion, los cuales ya son conocidos en la iteracion.
Output:
y - (tuple) El valor a y b, y el costo de calcularlo.
'''
def bisect_initial_guess(x_i, x_i_1, y_i):
    op_cost = 10 
    A = y_i
    B = (x_i_1 - x_i)/2
    C = x_i_1
    D = y_i*np.cos(x_i + y_i)
    up = (A + B*D)
    recta1 = up/(1 - B)
    recta2 = up/(1 + B)
    return recta1, recta2, op_cost

'''
Fixed point iteration.
Input:
x_i, x_i_1, y_i - (double) Parametros a evaluar en la funcion, los cuales ya son conocidos en la iteracion.
guess - (double) El initial guess a evaluar
tol - (double) La tolerancia del error.
N_budget - (integer) El presupuesto que nos queda para ejecutar este metodo.
Output:
y - (tuple) El valor y_i_1 que cumple el punto fijo y lo que resta del presupuesto.
'''
def fpi(x_i, x_i_1, y_i, guess, tol, N_budget):
    new_guess, aux_bud = g_aveng(x_i, x_i_1, y_i, guess)
    N_budget -= aux_bud
    while np.abs(new_guess - guess) > tol:
        N_budget -= 2
        guess = new_guess
        new_guess, aux_bud = g_aveng(x_i, x_i_1, y_i, guess)
        N_budget -= aux_bud
        if N_budget < 0:
            return np.nan, N_budget
    N_budget -= 2
    return new_guess, N_budget

'''
Metodo de la newton.
Input:
x_i, x_i_1, y_i - (double) Parametros a evaluar en la funcion, los cuales ya son conocidos en la iteracion.
guess - (double) El initial guess a evaluar
tol - (double) La tolerancia del error.
N_budget - (integer) El presupuesto que nos queda para ejecutar este metodo.
Output:
y - (tuple) El valor y_i_1 que vuelve cero la ecuacion y lo que resta de presupuesto.
'''
def newton(x_i, x_i_1, y_i, guess, tol, N_budget):
    guess_1 = guess
    result, aux = f_aveng(x_i, x_i_1, y_i, guess_1)
    N_budget -= aux
    while result > tol and N_budget > 0:
        a, aux = f_aveng(x_i, x_i_1, y_i, guess)
        N_budget -= aux
        b, aux = d_f_aveng(x_i, x_i_1, y_i, guess)
        N_budget -= aux
        guess_1 = guess - a/b
        N_budget -= 2
        result, aux = f_aveng(x_i, x_i_1, y_i, guess_1)
        N_budget -= aux
        if N_budget < 0:
            break
    return guess_1, N_budget
```

```python
def bisect_ode_solver(M_intervals=10,N_budget=1e4,tol=1e-8):
    y = np.full(M_intervals+1,np.nan)
    x = np.linspace(0,100,M_intervals+1)
    y[0] = 1.0
    for i in range(1, M_intervals+1):
        a, b, aux = bisect_initial_guess(x[i-1], x[i], y[i-1])
        N_budget -= aux
        aux_y, N_budget = bisect(x[i-1], x[i], y[i-1], a, b, tol, N_budget)
        if N_budget < 0:
            break
        y[i] = aux_y
    return y
def fpi_ode_solver(M_intervals=10,N_budget=1e4,tol=1e-8):
    y = np.full(M_intervals+1,np.nan)
    x = np.linspace(0,100,M_intervals+1)
    y[0] = 1.0
    for i in range(1, M_intervals+1):
        guess = 1
        aux_y, N_budget = fpi(x[i-1], x[i], y[i-1], guess, tol, N_budget)
        if N_budget < 0:
            break
        y[i] = aux_y
    return y
def newton_ode_solver(M_intervals=10,N_budget=1e4,tol=1e-8):
    y = np.full(M_intervals+1,np.nan)
    x = np.linspace(0,100,M_intervals+1)
    y[0] = 1.0
    for i in range(1, M_intervals+1):
        guess = bisect_initial_guess(x[i-1], x[i], y[i-1])[1]
        aux_y, N_budget = newton(x[i-1], x[i], y[i-1], guess, tol, N_budget)
        if N_budget < 0:
            break
        y[i] = aux_y
    return y
```

## 3.- Visualización y Conclusiones
Utilice el siguiente código para visualizar sus resultados. Se incluye esta función para que pueda visualizar sus resultados.

```python
def plot_solutions_of_ODE(M_intervals=10,N_budget=1e4,log_tol=-8,shift=0.1):
    tol = 10**log_tol
    x        = np.linspace(0,10,M_intervals+1)
    y_bisec  = bisect_ode_solver(M_intervals,N_budget,tol)
    y_fpi    = fpi_ode_solver(M_intervals,N_budget,tol)
    y_newton = newton_ode_solver(M_intervals,N_budget,tol)
    plt.figure(figsize=(8,8))
    plt.plot(x,y_bisec+shift,'b-',label='Bisección: y+'+str(shift))
    plt.plot(x,y_fpi,'m-',label='FPI: y')
    plt.plot(x,y_newton-shift,'g-',label='Newton: y-'+str(shift),linewidth=3,alpha=0.5)
    plt.legend(loc='best')
    plt.title('Plot of solutions with shift for visualization purposes')
    plt.grid(True)
    plt.show()
    
interact(plot_solutions_of_ODE,M_intervals=(100,100000,100),N_budget=(100,1000000,100),log_tol=(-14,-1,1),shift=(-1,1,0.01))
```
